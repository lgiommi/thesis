```python
from pandas import read_csv
import numpy as np
import pandas as pd
```

Load the two csv, the first one with tt simulated data and the second with real data

```python
filename = 'simulatoReduced.csv'
names=['category','mva','nBJets','nJets','nLeptons','jetEta_0','jetEta_1','jetEta_2','jetEta_3','jetEta_4',\
      'jetMass_0','jetMass_1','jetMass_2','jetMass_3','jetMass_4','jetMassSoftDrop_0','jetMassSoftDrop_1',\
      'jetMassSoftDrop_2','jetMassSoftDrop_3','jetMassSoftDrop_4',\
       'jetPhi_0','jetPhi_1','jetPhi_2','jetPhi_3','jetPhi_4',\
      'jetPt_0','jetPt_1','jetPt_2','jetPt_3','jetPt_4','jetTau1_0','jetTau1_1','jetTau1_2','jetTau1_3','jetTau1_4',\
      'jetTau2_0','jetTau2_1','jetTau2_2','jetTau2_3','jetTau2_4','jetTau3_0','jetTau3_1','jetTau3_2',\
      'jetTau3_3','jetTau3_4','triggerBit_0','triggerBit_1','triggerBit_2','triggerBit_3','triggerBit_4',\
      'triggerBit_5','triggerBit_6','triggerBit_7','triggerBit_8','triggerBit_9']
data_SIM = read_csv(filename, names=names)
```

```python
filename = 'reale3.5.csv'
data = read_csv(filename, names=names)
```

Select which cuts do you want to use for the dataset to train the model. Added a new column respect to the original dataset with 0 if data comes from real data dataset, otherwise if data comes from tt simulated dataset put 1.

```python
row_list=[]
for index, row in data.iterrows():
    if data['triggerBit_4'][index]==1 and data['nJets'][index]>=2 and \
    data['jetPt_0'][index]>400 and data['jetPt_1'][index]>400 and data['nLeptons'][index]==0 and \
    data['nBJets'][index]==0 and data['category'][index]==0:
        row_list.append(row.tolist())
print ("Fondo_DATI -> ", len(row_list))
fondo = pd.DataFrame(row_list,columns=names)
fondo['target']=np.zeros(len(row_list))
```

```python
row_list=[]
for index, row in data_SIM.iterrows():
    if data_SIM['triggerBit_4'][index]==1 and data_SIM['nJets'][index]>=2 and \
    data_SIM['jetPt_0'][index]>400 and data_SIM['jetPt_1'][index]>400 and data_SIM['nLeptons'][index]==0 and\
    data_SIM['nBJets'][index]==0 and data_SIM['category'][index]==0:
        row_list.append(row.tolist())
print ("Segnale_SIM -> ",len(row_list))
segnale = pd.DataFrame(row_list,columns=names)
segnale['target']=np.ones(len(row_list))
```

Concatenate the two dataframes and create the final dataset to train the model.

```python
frames=[segnale,fondo]
dataset = pd.concat(frames)
```

```python
dataset = dataset.sample(frac=1).reset_index(drop=True)
```
